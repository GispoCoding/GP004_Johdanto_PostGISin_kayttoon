Tässä osassa tutustutaan erilaisiin tietotyyppeihin.

Alusta notebook SQL-harjoituksia varten ([SQL-solujen ohje](H1_jupyter_tutuksi.ipynb#SQL-solujen-ajaminen)).

````python
# Lataa SQL-lisäosa
%load_ext sql
# Luo tietokantayhteys tietokantaan postgres
from ipygis import get_connection_url
c_url = get_connection_url()
%sql $c_url
````

Tutustutaan aluksi `tieliikenne`-taulun sisältöön

````python
%sql SELECT * FROM tieliikenne LIMIT 30;
````

Tarkistetaan, kuinka monen ajoneuvon ensirekiseröinti on tapahtunut vuoden 2015 jälkeen.

````python
%sql SELECT count(*) FROM tieliikenne WHERE EXTRACT(YEAR FROM ensirekisterointipvm) > 2015;
````

Yllä oleva solu tuottaa virheen: `function date_part(unknown, text) does not exist`. Tämä vihjaisi siis, että kentän ensirekisterointipvm tietotyyppi olisi `text`, mikä ei suinkaan ole paras tietotyyppi päivämäärän säilyttämiseen. Tarkistetaan vielä `information_schema`-taulusta pvm-kenttien tietotyypit:

````python
%%sql 
SELECT column_name, data_type FROM information_schema.columns 
WHERE table_name = 'tieliikenne' AND column_name LIKE '%pvm%';
````

Sarake ensirekisterointipvm tosiaan on tyypiltään `text` eikä yksi PostgreSQL:n aikatyypeistä (`date`, `interval`, `timestamp`...). Yksi ratkaisu ongelmaan olisi muuttaa sarakkeen tyyppi kyselyssä oikeaan muotoon:

````python
%sql SELECT count(*) FROM tieliikenne WHERE date_part('year', to_date(ensirekisterointipvm, 'YYYY-MM-DD')) > 2015;
````

Tietotyyppimuunnoksen (castaus) voi tehdä myös käyttäen `::`-operaattoria:

````python
%sql SELECT count(*) FROM tieliikenne WHERE extract(year from ensirekisterointipvm::date) > 2015;
````

Tässä tapauksessa fiksumpaa olisi kuitenkin muuttaa koko sarakkeen tyyppi oikeaksi, sillä kyseessä on selkeästi päivämääräsarake.

## **Harjoitus 4.1**: Ota oikea tietotyyppi käyttöön
Muuta `tieliikenne`-taulun  ensirekisterointipvm-sarake käyttämään oikeaa `date`-tietotyyppiä.
Käytä syntaksia:

```sql
ALTER TABLE taulun_nimi ALTER COLUMN kenttä TYPE <tyyppi>
using to_date(kenttä, 'päivämääräformaatti');
```

Voit tarkistaa ohjeen `using_date`-funktion käyttämiseen ja päivämääräformaatin muotoon alla olevasta upotuksesta:

````python
from IPython.display import IFrame, display
r = %sql SELECT ensirekisterointipvm, kayttoonottopvm FROM tieliikenne LIMIT 10
display(r.DataFrame())
IFrame('https://www.postgresql.org/docs/13/functions-formatting.html#FUNCTIONS-FORMATTING-DATETIME-TABLE', width=900, height=350)
````

Kirjoita kentän muuttamisen lause alla olevaan soluun. Muunnoksen suorittamisessa menee hetki.

````python
%%sql

ALTER ...
````

````python
%%sql 

ALTER TABLE tieliikenne ALTER COLUMN ensirekisterointipvm TYPE DATE 
using to_date(ensirekisterointipvm, 'YYYY-MM-DD')
````

### Tarkistus
Kokeile nyt uudestaan tiedoston alussa olevaa kyselyä:

````python
res = %sql SELECT count(*) FROM tieliikenne WHERE date_part('year', ensirekisterointipvm) > 2015;
assert res[0][0] == 851757, f'Tarkasta muunnos, sait {res[0][0]} riviä vastaukseksi'
````

## Geometriatyyppi
PostGIS-lisäosan mukana tulee laaja tuki erilaisille geometriatyypeille ja muunnosfunktioille.

Katso taulun `kunnat` dataa Helsinki-rivin kautta.

````python
%sql SELECT * FROM kunnat WHERE nimi = 'Helsinki'; 
````

geom-kentän sisältö on binääristä, eikä sitä pysty tulkitsemaan. PostGISin `ST_AsText`-funktiolla tämä onnistuu

````python
%sql SELECT nimi, ST_AsText(geom) FROM kunnat WHERE nimi = 'Helsinki'; 
````

Nyt paljastuu jo, että kyseessä on polygoni, tai tarkemmin multipolygoni, joka sisältää paljon koordinaatteja. Harjoitusympäristöön on rakennettu toiminnalisuus, jolla geometriakohteita pystyy visualisoimaan interaktiivisesti

````python
from ipygis import get_map
rs = %sql SELECT * FROM kunnat WHERE nimi = 'Helsinki';
map = get_map(rs)
map
````
